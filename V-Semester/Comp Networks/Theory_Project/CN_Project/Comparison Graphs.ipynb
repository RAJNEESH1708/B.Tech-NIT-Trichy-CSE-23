# Comparisons between Proposed and Other Hashing Algorithms for IP Address Lookup in Computer Networks

```python
import math
import csv
import operator as op
from functools import reduce
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("A_performance.csv")
csvreader = csv.reader(file)
header = next(csvreader)
print(header)
db = []
for row in csvreader:
    db.append(row[1:])
print(db)
file.close()
```

```python
for i in range(len(db)):
    for j in range(len(db[i])):
        db[i][j]=float(db[i][j])
```

## 1) Performance Comparison by ASL on Real IP Addresses

```python
# Using Numpy to create an array X
X = [2,3,4,5,6,7,8]

tlist = list(zip(*db))
# Assign variables to the y axis part of the curve

asl1 = tlist[0]
asl2 = tlist[1]
asl3 = tlist[2]
asl4 = tlist[3]

# Plotting both the curves simultaneously
plt.axis([2, 8, 0, 10])
plt.plot(X, asl1, color='r', label='ASL')
plt.plot(X, asl2, color='g', label='Linear ASL')
plt.plot(X, asl3, color='b', label='Bit Extraction ASL')
plt.plot(X, asl4, color='y', label='Group XORing ASL')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("m")
plt.ylabel("Average Search Length (ASL)")
plt.title("Performance Comparison with ASL on Artificial IP Addresses")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```

## 2) Performance Comparison by MSL on Real IP Addresses

```python
# Using Numpy to create an array X
X = [2,3,4,5,6,7,8]

# Assign variables to the y axis part of the curve

msl1 = tlist[4]
msl2 = tlist[5]
msl3 = tlist[6]
msl4 = tlist[7]

# Plotting both the curves simultaneously
plt.axis([2, 8, 0, 20])
plt.plot(X, msl1, color='r', label='MSL')
plt.plot(X, msl2, color='g', label='Linear MSL')
plt.plot(X, msl3, color='b', label='Bit Extraction MSL')
plt.plot(X, msl4, color='y', label='Group XORing MSL')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("m")
plt.ylabel("Maximum Search Length (MSL)")
plt.title("Performance Comparison with MSL on Artificial IP Addresses")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```

## 3) Performance Gain over the Standard Group XOR on Real IP Addresses

```python
# Using Numpy to create an array X
X = [2,3,4,5,6,7,8]

# Assign variables to the y axis part of the curve

asl_adhoc_pivot=list(tlist[0])
asl_group_XOR=list(tlist[3])
gain1=[]
for i in range(7):
    gain1.append((float(asl_group_XOR[i])-float(asl_adhoc_pivot[i]))/float(asl_group_XOR[i])*100.0)

# Plotting both the curves simultaneously
plt.axis([2, 8, -200, 200])
plt.plot(X, gain1, color='b', label='Adhoc Pivot Algorithm')
plt.plot(X, [0]*7, '-.',color='r')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("m")
plt.ylabel("% Gain")
plt.title("Performance Gain over the Standard Group XOR on Artificial IP Addresses")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```

## 4) Performance Gain over Advanced Bit Extraction Algorithm on Real IP Addresses

```python
# Using Numpy to create an array X
X = [2,3,4,5,6,7,8]

# Assign variables to the y axis part of the curve

asl_bit=list(tlist[2])
gain2=[]
for i in range(7):
    gain2.append((float(asl_bit[i])-float(asl_adhoc_pivot[i]))/float(asl_bit[i])*100.0)

# Plotting both the curves simultaneously
plt.axis([2, 8, -200, 200])
plt.plot(X, gain2, color='g', label='Adhoc Pivot Algorithm')
plt.plot(X, [0]*7, '-.',color='r')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("m")
plt.ylabel("% Gain")
plt.title("Performance Gain over Advanced Bit Extraction Algorithm on Artificial IP Addresses")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
